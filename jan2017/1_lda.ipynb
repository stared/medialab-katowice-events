```python
df = pd.read_csv("/Users/pmigdal/Downloads/events.csv")
```

```python
opisy_kol = df["description"].dropna()
```

```python
opisy_kol.iloc[0]
```

```python
# lepsza tokenizacja sie przyda
opisy = [opis.split() for opis in opisy_kol]
```

```python
len(opisy)
```

```python
dictionary = corpora.Dictionary(opisy)
dictionary.filter_extremes(no_below=5, no_above=0.2, keep_n = None)
corpus = [dictionary.doc2bow(opis) for opis in opisy]
```

```python
%time lda = models.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=2)
```

```python
lda
```

```python
lda2vis = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
```

```python
pyLDAvis.display(lda2vis)
```
