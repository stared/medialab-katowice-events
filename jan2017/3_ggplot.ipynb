```r
library(ggplot2)
library(dplyr)
```

```r
events <- read.csv("data/events.csv")
```

```r
colnames(events)
```

```r
events <- events %>% filter(attending_count > 1)
```

```r
ggplot(events, aes(attending_count, maybe_count)) +
  geom_point() +
  scale_x_log10() + 
  scale_y_log10()
```

```r
ggplot(events, aes(attending_count, maybe_count, color=attending_count/(attending_count+noreply_count), alpha=declined_count/(noreply_count+declined_count))) +
  geom_point() +
  scale_x_log10() + 
  scale_y_log10()
```
