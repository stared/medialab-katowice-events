```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
pairs = pd.read_csv("data/processed/pair_count.csv", index_col=[0, 1])
```

```python
pairs.head()
```

```python
events = pd.read_csv("data/events.csv", index_col='facebook_id')
events = events.loc[:, ['facebook_page', 'name', 'place_name', 'owner_id', 'owner_name']]
```

```python
def show_related(event_id=None, limit=10, by='pmi', min_count=2):
    if event_id is None:
        event_id = events.sample().index[0]
    print(events.loc[event_id])
    related = pairs.loc[event_id].loc[lambda df: df['count'] >= min_count] \
      .sort_values(by=by, ascending=False).head(limit)
    return related.join(events, how='inner')
```

```python
show_related()
```

```python
show_related(by='count')
```
