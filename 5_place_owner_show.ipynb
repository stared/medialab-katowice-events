```python
%matplotlib inline
import pandas as pd
import numpy as np
```

## Owners

```python
coowners = pd.read_csv("data/processed/pair_count_maybe_owner.csv", index_col=[0, 1])
```

```python
def show_related(pair_df, name=None, by='pmi', limit=10):
    if not name:
        name = pd.Series(pair_df.index.levels[0]).sample(1).iloc[0]
    print(name)
    res = pair_df.loc[name].copy()
    res['ratio'] = res['count'] / res['count_y']
    res = res.sort_values(by=by, ascending=False).head(limit)
    return res
```

```python
show_related(coowners)
```

```python
show_related(coowners, by='count')
```

## Places

```python
coplaces = pd.read_csv("data/processed/pair_count_maybe_place.csv", index_col=[0, 1])
```

```python
show_related(coplaces)
```

```python
show_related(coplaces, by='count')
```
