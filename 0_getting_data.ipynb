```python
import pandas as pd
import json
```

```python
import pymysql
# https://pymysql.readthedocs.io/en/latest/user/examples.html

pymysql.install_as_MySQLdb()
# crucial to make it compatible with Panas
```

```python
with open("mysql_credentials.json") as f:
    credentials = json.load(f)
```

```python
connection = pymysql.connect(host=credentials['host'],
                             user=credentials['user'],
                             password=credentials['password'],
                             db='events-tag')
```

```python
with connection.cursor() as cursor:
    cursor.execute("show tables;")
    result = cursor.fetchall()
    print(result)
```

```python
connection.close()
```

```python
mysql_uri_template = "mysql://{user}:{password}@{host}/{db}?charset=utf8mb4"
```

```python
db_string = mysql_uri_template.format(**credentials, db="events-tag")

pd.read_sql_table('events_tags', db_string).head()
```
